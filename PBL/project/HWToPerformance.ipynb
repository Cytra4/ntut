```python
# ================================================
#  項目 3：作業 → 課堂表現（有序邏輯斯回歸）
# ================================================

import pandas as pd
import statsmodels.api as sm
from statsmodels.miscmodels.ordinal_model import OrderedModel

#-------------------------------------------------
# 1. 讀取資料
#-------------------------------------------------
df = pd.read_csv("cleaned_merged_ch.csv", encoding="utf-8-sig", low_memory=False)

#-------------------------------------------------
# 2. 定義自變數（作業類型 + 作業數量）
#-------------------------------------------------

# 作業類型（國文)
hw_types_chi = [
    "做習題標準化_國文",
    "作文練習標準化_國文",
    "剪報或蒐集資料標準化_國文",
    "聽演講、參觀展覽或看表演標準化_國文",
    "分組專題活動標準化_國文"
]

# 作業數量（整體負荷）
hw_amount = "作業經常次數_國文"

X_cols = hw_types_chi + [hw_amount]

#-------------------------------------------------
# 3. 定義依變數
#-------------------------------------------------

y_vars = {
    "跟得上進度": "跟得上進度_國文",
    "用功程度": "用功程度_國文",
    "作業表現": "作業表現_國文"
}

#-------------------------------------------------
# 4. 將老師評量轉成「數值越大 = 表現越好」
#-------------------------------------------------

def recode_teacher_rating(series, reverse=False):
    """
    reverse=True 表示原始數值越小越好（如作業遲交）
    """
    if reverse:
        return series.replace({
            1: 1,  # 總是遲交
            2: 2,
            3: 3,
            4: 4   # 從未遲交（最好）
        })
    else:
        return series.replace({
            1: 4,  # 非常好
            2: 3,
            3: 2,
            4: 1   # 非常不好
        })

#-------------------------------------------------
# 5. 逐一跑三個有序邏輯斯回歸
#-------------------------------------------------

for label, y_col in y_vars.items():

    temp_df = df[X_cols + [y_col]].dropna()

    # 重新編碼 y
    if label == "作業表現":
        y = recode_teacher_rating(temp_df[y_col], reverse=True)
    else:
        y = recode_teacher_rating(temp_df[y_col])

    X = temp_df[X_cols]

    model = OrderedModel(
        y,
        X,
        distr="logit"
    )

    result = model.fit(method="bfgs", disp=False)

    print("\n" + "="*50)
    print(f" 國文：作業 → {label}")
    print("="*50)
    print(result.summary())
```
